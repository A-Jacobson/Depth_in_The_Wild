```python
loss_log = lambda target, pred: np.log(1 + np.exp(-target * pred))
loss_square = lambda target: target **2
```

```python
preds = np.linspace(-4, 4)
target_1 = np.ones_like(preds)
target_n1 = np.ones_like(preds) * -1

loss_1 = loss_log(target_1, preds)
loss_n1 = loss_log(target_n1, preds)
loss_0 = loss_square(preds)
```

```python
plt.plot(preds, loss_1)
plt.vlines(1, 0, 4, colors='Red', linestyles='dashed')
plt.xlabel('predictions')
plt.ylabel('loss')
plt.title('loss when target = 1');
```

```python
plt.plot(preds, loss_n1)
plt.vlines(-1, 0, 4, colors='Red', linestyles='dashed')
plt.xlabel('predictions')
plt.ylabel('loss')
plt.title('loss when target = -1');
```

```python
plt.plot(preds, loss_0)
plt.vlines(0, -1, 16, colors='Red', linestyles='dashed')
plt.xlabel('predictions')
plt.ylabel('loss')
plt.title('loss when target = 0');
```
